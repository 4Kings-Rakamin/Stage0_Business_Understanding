###  1. Berapa perbandingan jumlah kandidat yang diterima dan tidak ?

```python
sns.countplot(x='HiringDecision', data=df,hue='HiringDecision', palette='viridis')
plt.title('Hiring Decision Distribution')
plt.show()
```

```python
total_HD=df['HiringDecision'].value_counts()
total_HD
```

```python
labels_pie = [ 'Ditolak', 'Diterima',]
plt.figure(figsize=(6,6))
plt.pie(total_HD,
        labels=labels_pie,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Presentase Keputusan Perekrutan")
plt.show()
```

### 2.Apakah InterviewScore tinggi, peluang diterima lebih besar?

```python
sns.barplot(x='HiringDecision',y='InterviewScore', hue='HiringDecision', data=df, palette='viridis')
plt.title('Keputusan Perekrutan berdasarkan Skor Wawancara')
plt.show()
```

### 3.Apakah DistanceFromCompany memengaruhi keputusan hiring?

### 4. Apakah EducationLevel berperan besar dalam keputusan hiring ?

```python
sns.countplot(x='EducationLevel',hue='HiringDecision', data=df, palette='viridis')
plt.title('Keputusan Perekrutan berdasarkan Tingkat Pendidikan')
plt.show()
```

### 5. berapa perbandingan pria dan wanita yang diterima dan tidak ?

## Univariate Analysis

```python
numeric= df.select_dtypes(include=['int64','float64']).columns
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(7, 9))
axes = axes.flatten()

for i, col in enumerate(numeric):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(col)

# Hide any unused subplots
for j in range(len(numeric), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(10, 5))
axes = axes.flatten()

for i, col in enumerate(numeric):
    sns.kdeplot(df[col], ax=axes[i], fill=True, color='skyblue')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')

for j in range(len(numeric), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

## Multivariate Analisis
